```python
import matplotlib.pyplot as plt
import numpy as np

# Utility to label bars
def autolabel(ax, rects, fmt="{:.0f}%"):
    for r in rects:
        height = r.get_height()
        ax.annotate(fmt.format(height),
                    xy=(r.get_x() + r.get_width() / 2, height),
                    xytext=(0, 6), textcoords="offset points",
                    ha='center', va='bottom')

# Gradient palette (consistent with your team)
blue_gradient = ['#0b3d91', '#1e5bbf', '#3a7bd5', '#5ca9e6', '#8ac9f9']
bg_light_blue = '#eaf4fc'

# ======================================================
# Chart A — Shopping Experience & Digital Assistants
# ======================================================
fig, ax = plt.subplots(figsize=(9, 5))
fig.patch.set_facecolor(bg_light_blue)
ax.set_facecolor(bg_light_blue)

labels = ["Total\nRespondents", "Among\nParents", "Agree", "Disagree"]
values = [69, 80, 54, 25]
x = np.arange(len(labels))
width = 0.6

# Bars
bars = ax.bar(x, values, width, color=[blue_gradient[2], bg_light_blue, blue_gradient[3], blue_gradient[4]],
              edgecolor=blue_gradient[0], linewidth=1.5)

autolabel(ax, bars)

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 90)
ax.set_ylabel("Percentage")
ax.set_title("Shopping Experience & Digital Assistants — Key Findings")

ax.grid(axis='y', linestyle=':', linewidth=0.8, color=blue_gradient[4])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()

# ======================================================
# Chart B1 — Trust Split (AI vs Influencers vs Don’t Know)
# ======================================================
fig, ax = plt.subplots(figsize=(5.5, 5))
fig.patch.set_facecolor(bg_light_blue)
ax.set_facecolor(bg_light_blue)

categories = ["Trust AI-based\nrecommendations",
              "Trust social media\ninfluencers",
              "Don't know"]
vals = [27, 24, 49]

bars = ax.barh(range(len(categories)), vals,
               color=[blue_gradient[2], blue_gradient[0], blue_gradient[4]],
               edgecolor=blue_gradient[0], linewidth=1.2, height=0.7)

for i, bar in enumerate(bars):
    ax.text(vals[i] + 2, bar.get_y() + bar.get_height() / 2, f"{vals[i]}%",
            va='center', ha='left', fontsize=11)

ax.set_yticks(range(len(categories)))
ax.set_yticklabels(categories)
ax.invert_yaxis()
ax.set_xlim(0, 60)
ax.set_xlabel("Percentage")
ax.set_title("Individuals are split on who to trust")

ax.grid(axis='x', linestyle=':', linewidth=0.8, color=blue_gradient[4])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()

# ======================================================
# Chart B2 — Trust by Segment (AI vs Influencers)
# ======================================================
segments = ["Income $100K", "Males", "Females", "GenXers", "Parents", "Walmart regular\nshoppers"]
ai_vals = [40, 36, 26, 32, 39, 30]
influencer_vals = [23, 23, 18, 20, 30, 27]

x = np.arange(len(segments))
w = 0.35

fig, ax = plt.subplots(figsize=(10, 5.2))
fig.patch.set_facecolor(bg_light_blue)
ax.set_facecolor(bg_light_blue)

bars1 = ax.bar(x - w/2, ai_vals, w, label="AI-based", color=blue_gradient[2], edgecolor=blue_gradient[0])
bars2 = ax.bar(x + w/2, influencer_vals, w, label="Social media influencers", color=blue_gradient[0], edgecolor=blue_gradient[0])

autolabel(ax, bars1)
autolabel(ax, bars2)

ax.set_xticks(x)
ax.set_xticklabels(segments)
ax.set_ylim(0, 50)
ax.set_ylabel("Percentage")
ax.set_title("Trust: AI-based recommendations vs social media influencers")
ax.legend(frameon=False, loc='upper left')

ax.grid(axis='y', linestyle=':', linewidth=0.8, color=blue_gradient[4])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Your theme palette (blue gradient + optional orange accent) ---
blue_gradient = ['#0b3d91', '#1e5bbf', '#3a7bd5', '#5ca9e6', '#8ac9f9']  # dark → light
accent_orange = '#f7933f'
bg_light = '#eaf4fc'

def label_inside_pie(ax, pct, txt, **kw):
    ax.text(0, 0, f"{pct}%\n{txt}", ha='center', va='center', fontsize=18, **kw)

# ===============================================================
# 1) Pie — “How much do you plan to invest in generative AI?”
#    (use blue shades instead of NVIDIA green)
# ===============================================================
labels = ["< $5M", "$5–15M", "$16–30M", "$31–50M", "$51–100M", "$0"]
vals   = [77,      9,        5,         5,          2,           3]
colors = [blue_gradient[2], blue_gradient[0], blue_gradient[1],
          blue_gradient[3], blue_gradient[4], "#d9e7fb"]

fig, ax = plt.subplots(figsize=(6,6))
fig.patch.set_facecolor(bg_light); ax.set_facecolor(bg_light)

wedges, _ = ax.pie(vals, colors=colors, startangle=90, wedgeprops=dict(edgecolor=blue_gradient[0], linewidth=1))
label_inside_pie(ax, 77, "Less than\n$5M", color=blue_gradient[0])

# Legend styled like the original keys
legend_labels = [f"{labels[i]}: {vals[i]}%" for i in range(len(labels))]
ax.legend(wedges, legend_labels, loc="lower center", bbox_to_anchor=(0.5, -0.05), ncol=2, frameon=False)
ax.set_title("How much do you plan to invest in generative AI?")
plt.tight_layout()
plt.show()

# ===============================================================
# 2) Horizontal bars — “Which channel provides biggest revenue growth?”
#    (two selections possible)
# ===============================================================
cats = [
    "Ecommerce (company website)",
    "Brick-and-mortar stores",
    "Mobile application",
    "Third-party ecommerce (e.g., Amazon)",
    "Social media",
    "Web 3.0/metaverse for enterprises",
    "Web 3.0/metaverse for consumers",
]
vals = [70, 30, 26, 17, 13, 4, 4]
colors = [blue_gradient[2]] + [blue_gradient[3]]*2 + [blue_gradient[4]]*3 + [blue_gradient[4]]

fig, ax = plt.subplots(figsize=(8,4.6))
fig.patch.set_facecolor(bg_light); ax.set_facecolor(bg_light)

y = np.arange(len(cats))
bars = ax.barh(y, vals, color=colors, edgecolor=blue_gradient[0], linewidth=1.2, height=0.7)
ax.set_yticks(y, cats)
ax.invert_yaxis()
ax.set_xlim(0, 80)
ax.set_xlabel("Share of respondents (%)")
ax.set_title("What channel provides the biggest opportunity for revenue growth?\n(two selections possible)")
for i, b in enumerate(bars):
    ax.text(b.get_width()+1, b.get_y()+b.get_height()/2, f"{vals[i]}%", va='center')
for sp in ax.spines.values(): sp.set_visible(False)
ax.grid(axis='x', linestyle=':', color=blue_gradient[4])
plt.tight_layout()
plt.show()

# ===============================================================
# 3) Horizontal bars — “Areas to use gen AI to transform business”
# ===============================================================
areas = [
    "Customer experiences",
    "Marketing and content\ngeneration",
    "Employee experiences",
    "Product design and onboarding",
    "Other"
]
vals = [86, 68, 57, 46, 14]
colors = [blue_gradient[2], blue_gradient[2], blue_gradient[3], blue_gradient[3], blue_gradient[4]]

fig, ax = plt.subplots(figsize=(9,3.6))
fig.patch.set_facecolor(bg_light); ax.set_facecolor(bg_light)

y = np.arange(len(areas))
bars = ax.barh(y, vals, color=colors, edgecolor=blue_gradient[0], linewidth=1.2, height=0.7)
ax.set_yticks(y, areas)
ax.invert_yaxis()
ax.set_xlim(0, 100)
ax.set_xlabel("Share of respondents (%)")
ax.set_title("What are the areas you may use generative AI in to transform your business?\n(all retail respondents)")
for i, b in enumerate(bars):
    ax.text(b.get_width()+1, b.get_y()+b.get_height()/2, f"{vals[i]}%", va='center')
for sp in ax.spines.values(): sp.set_visible(False)
ax.grid(axis='x', linestyle=':', color=blue_gradient[4])
plt.tight_layout()
plt.show()

# ===============================================================
# 4) Pie — “How many gen-AI use cases are you considering / implemented?”
#    three-slice pie with callout-like labels
# ===============================================================
labels = ["1–2 use cases", "3–5 use cases", "6+ use cases"]
vals   = [18, 25, 57]
colors = [blue_gradient[4], blue_gradient[3], blue_gradient[2]]

fig, ax = plt.subplots(figsize=(7,4.2), layout='constrained')
fig.patch.set_facecolor(bg_light); ax.set_facecolor(bg_light)

wedges, texts = ax.pie(vals, colors=colors, startangle=90,
                       wedgeprops=dict(edgecolor=blue_gradient[0], linewidth=1))
ax.set_title("How many generative AI use cases are you considering or have already implemented?")

# Callout-style labels roughly matching the screenshot look
kw = dict(ha="center", va="center", fontsize=12)
# Place labels outside with simple lines
for i, w in enumerate(wedges):
    ang = (w.theta2 + w.theta1)/2
    x = np.cos(np.deg2rad(ang))
    y = np.sin(np.deg2rad(ang))
    ax.plot([0, 1.2*x], [0, 1.2*y], color=blue_gradient[0], lw=1)
    ax.text(1.35*x, 1.35*y, f"{vals[i]}%\n{labels[i]}", **kw)

# Optional subtle watermark-style center text (comment out if not needed)
# ax.text(0, 0, "AI", ha='center', va='center', fontsize=16, color=blue_gradient[0], alpha=0.15)

plt.show()
```
