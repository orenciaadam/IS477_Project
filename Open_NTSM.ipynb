```python
df_clean = df[
    (df['imdb_id'].notna()) &
    (df['imdb_id'].astype(str).str.len() >= 7)
].copy()

df_clean['imdb_id'] = df_clean['imdb_id'].astype(str).str.strip()
```

```python
df_clean
```

```python
movies = df_clean[df_clean["type"] == "MOVIE"]
shows = df_clean[df_clean["type"] == "SHOW"]
unknown = df_clean[~df_clean["type"].isin(["MOVIE", "SHOW"])]
print(f"Number of movies: {len(movies)}")
print(f"Number of shows: {len(shows)}")
print(f"Number of unknown: {len(unknown)}")
```
